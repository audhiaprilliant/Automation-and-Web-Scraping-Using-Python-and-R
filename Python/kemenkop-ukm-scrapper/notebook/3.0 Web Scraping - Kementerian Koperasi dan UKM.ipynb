## Import modules

```python
# Module for JSON
import json
# Module for folders
import os
```

## Read the data

```python
# Base path
base_path = '../data/raw'
  
# Get all directories in our base path.
folder = os.listdir(base_path)

# List all folders
print('Before moving directories:')
print(sorted(os.listdir(base_path), key = lambda x: int(x[0:2])))
```

```python
'Kab. Simeulue - District.json'.replace(' - District.json', '')
```

```python
# Traverse each folders in base path
for dir_ in folder:
    # Subfolders
    subfolders = os.path.join(base_path, dir_)
    print('=============')
    # Files in subfolders
    subsubfolders = os.listdir(subfolders)
    for subsubfolder in subsubfolders:
        subsubsufolders = os.path.join(subfolders, subsubfolder)
        # Files in subfolders
        files = os.listdir(subsubsufolders)
        for file in files:
            print(subsubsufolders)
            data_dir = os.path.join(subsubsufolders, file)
            # OPEN JSON FILE
            with open(data_dir) as f:
                file_json = json.load(f)
            links = file_json['Links']
            for link in links:
                print(link)
```
